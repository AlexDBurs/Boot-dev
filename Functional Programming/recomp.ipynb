````python
def stylize_title(document):
    return add_border(center_title(document))


# Don't touch below this line


def center_title(document):
    width = 40
    title = document.split("\n")[0]
    centered_title = title.center(width)
    return document.replace(title, centered_title)


def add_border(document):
    title = document.split("\n")[0]
    border = "*" * len(title)
    return document.replace(title, title + "\n" + border)
````

````python
def file_type_getter(file_extension_tuples):
    file_ext_dict = {}
    for filetype, extensions in file_extension_tuples:
        for extension in extensions:
            file_ext_dict[extension] = filetype
    return lambda extension: file_ext_dict.get(extension, "Unknown")

````

````python
def change_bullet_style(document):
    line_list = document.split(sep = "\n")
    new_bullet_line_list = map(convert_line,line_list)
    return "\n".join(new_bullet_line_list)


# Don't edit below this line


def convert_line(line):
    old_bullet = "-"
    new_bullet = "*"
    if len(line) > 0 and line[0] == old_bullet:
        return new_bullet + line[1:]
    return line
````

````python
import functools


def join(doc_so_far, sentence):
    return doc_so_far + ". " + sentence


def join_first_sentences(sentences, n):
    if n==0:
        return ""
    sentences_list = sentences[:n]
    return functools.reduce(join, sentences_list) + "."

````

````python
valid_formats = [
    "docx",
    "pdf",
    "txt",
    "pptx",
    "ppt",
    "md",
]

# Don't edit above this line


def pair_document_with_format(doc_names, doc_formats):
    doc_list = zip(doc_names, doc_formats)
    is_doc_format_valid = lambda doc: doc[1] in valid_formats
    return list(filter(is_doc_format_valid, doc_list))
````

````python
def zipmap(keys, values):
    if len(keys) == 0 or len(values)==0:
        return {}
    
    zip_dict = {keys[0]: values[0], **zipmap(keys[+1:],values[+1:])}

    return zip_dict

````

````python
def sum_nested_list(lst):
    if len(lst) <= 0:
        return 0
    total = 0
    for l in lst:
        if isinstance(l, list):
            total += sum_nested_list(l)
        else:
            total += l
        
    return total
````

````python
def list_files(parent_directory, current_filepath=""):
    dir_list = []
    for child_dir in parent_directory.keys():
        current_dir = f'{current_filepath}/{child_dir}'
        if parent_directory[child_dir] == None:
            dir_list.append(current_dir)
        else:
            dir_list.extend(list_files(parent_directory[child_dir], current_dir))
    return dir_list
````

````python
def find_longest_word(document, longest_word=""):

    doc_list =document.split(maxsplit = 1)
    
    if len(doc_list) == 0:
        return longest_word        

    if len(doc_list[0]) > len(longest_word):
        longest_word = doc_list[0]

    if len(doc_list) > 1:
        return find_longest_word(doc_list[1], longest_word)
    
    return longest_word
````

````python
def count_nested_levels(nested_documents, target_document_id, level=1):
    for document_id in nested_documents:
        if document_id == target_document_id:
            return level
        
        else:
            nested_level = count_nested_levels(nested_documents[document_id], target_document_id, level + 1)
            if nested_level != -1:
                return nested_level
            
    return -1
````
